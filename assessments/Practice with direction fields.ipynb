```python
# Now put in the arrows
# The locations are at X, Y shown above
# Need to specify the directions
# Those are given by the DE
# Let's plot the field for dy/dx = y^2 - x

dy = Y - X
dx = np.ones(dy.shape)  # Returns an array of all 1's, same length as dy

plt.quiver(X,Y,dx,dy,color = "red")
plt.show()
```

```python
# Issues: 
# Arrows are different lengths so let's rescale so that they are unit vectors

norm = np.sqrt(dx**2 + dy**2)
dyu = dy/norm
dxu = dx/norm

plt.quiver(X,Y,dxu,dyu,color="purple")
plt.show()
```

```python
# Putting it all together 

x = np.arange(0,4,.3)
y = np.arange(0,10,.3)

X, Y = np.meshgrid(x,y)

dy = Y - X
dx = np.ones(dy.shape)
norm = np.sqrt(dx**2 + dy**2)
dyu = dy/norm
dxu = dx/norm

plt.quiver(X,Y,dxu,dyu,color="purple")
plt.show()
```

```python
x = np.arange(0,2,.3)
y = np.arange(-5,5,.3)

X, Y = np.meshgrid(x,y)

dy = Y - X
dx = np.ones(dy.shape)
norm = np.sqrt(dx**2 + dy**2)
dyu = dy/norm
dxu = dx/norm


plt.quiver(X,Y,dxu,dyu,color="purple")

soln = x - 0.303265329856316*np.exp(x) + 1
plt.plot(x,soln)
plt.show()
```

```python
# Trying a direction field for a system: 
# dx/dt = -x + 6y, dy/dt = - x

x = np.arange(-4,5,.5)
y = np.arange(-3,4,.5)

X, Y = np.meshgrid(x,y)

dy = -X
dx = -X + 6*Y
norm = np.sqrt(dx**2 + dy**2)
dyu = dy/norm
dxu = dx/norm

plt.quiver(X,Y,dxu,dyu,color="purple")
plt.show()
```

```python
# Trying a direction field for a system: 
# dx/dt = -x + 6y, dy/dt = - x

x = np.arange(-5,6,.5)
y = np.arange(-1,6,.5)

X, Y = np.meshgrid(x,y)

dy = -5/2*X + 7/4*Y
dx = -3/4*X - 15/8*Y
norm = np.sqrt(dx**2 + dy**2)
dyu = dy/norm
dxu = dx/norm

plt.quiver(X,Y,dxu,dyu,color="purple")
plt.show()
```
