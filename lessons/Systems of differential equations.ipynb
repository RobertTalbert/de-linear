# Introduction to Systems of Differential Equations

## MTH 302: Linear Algebra and Differential Equations

This tutorial is a simplified introduction to **systems of differential equations**. It goes with Sections 3.1 and 3.2 in your textbook. 

### What is a system of differential equations? 

Imagine you have two populations in an ecosystem, where one population feeds off of the other, like wolves and rabbits. If it weren't for the wolves, the rabbit population would grow without restriction. And if it weren't for the rabbits, the wolf population would go extinct because the rabbits are the sole source of food for the wolves. 

In other words, without the predators, the rabbit population would **grow over time at a rate proportional to its size**. If we let $R(t)$ be the rabbit population at time $t$,   it would satisfy a differential equation that looks like 
$$\frac{dR}{dt} = aR$$
where $a$ is some positive constant. And without the rabbits, the wolf population would decay over time at a rate propotional to its size. If $W(t)$ is the wolf population at time $t$, then 
$$\frac{dW}{dt} = dW$$
where $d$ is some other constant (a negative constant, since the population is decreasing).

However, this assumes that the wolves and rabbits never interact -- which is definitely not the case here. The wolves eat the rabbits and the rabbits are eaten by the wolves, so **the growth rate of each population depends not only on its own population size but the *other* population's size.** So there should be **two variables** on each side of these differential equations: 

$$\begin{align}
\frac{dR}{dt} &= aR + bW \\
\frac{dW}{dt} &= cR + dW
\end{align}$$

In other words, the growth rate in the rabbit population is affected not only by its own size but also the size of the wolf population. (More wolves, slower growth rate for rabbits.) And likewise the growth of the wolf population is impacted not only by its own size but the number of rabbits too. (More rabbits, higher growth rate for wolves.) 

This is a **system of differential equations**: Two equations, involving two different functions of time, where both deriviatives are connected to the values of both functions. Compare this to a single differential equation, which relates a function to its derivative; here we have two functions, each of which is related to its own derivative as well as the other's derivative. 

**Example:** Here's an instance of the predator-prey system described above: 

$$\begin{align}
\frac{dR}{dt} &= 4.5R - 2.5W \\
\frac{dW}{dt} &= 7.5R  - 5.5W
\end{align}$$

This would say that if there were no wolves ($W = 0$) then the rabbit population would undergo exponential growth ($\frac{dR}{dt} = 4.5R$) and if there were no rabbits ($R = 0$) then the wolf population would undergo exponential decay ($\frac{dW}{dt} = -5.5W$). But, the presence of wolves reduces the growth rate of the rabbits; and the presence of rabbits increases the growth rate of the wolves. 

(Note: These numbers are wildly out of proportion. In real populations, those parameters are between 0 and 1. The numbers you see here are used as a hypothetical example, to make some of the upcoming math work out nicely.) 

### What is a solution to a system of differential equations? 

Just like the solution to a single differential equation is a *function* that satisfies the DE, a solution to a system of DE's is **a pair of functions that satisfy both DE's simultaneously**. (If this is beginning to sound like linear algebra, you are on the right track.)

Take the system of DE's above for instance: 
$$\begin{align}
\frac{dR}{dt} &= 4.5R - 2.5W \\
\frac{dW}{dt} &= 7.5R  - 5.5W
\end{align}$$

The pair of functions $R(t) = e^{2t}, W(t) = e^{2t}$ is a solution to this system. To see this, look at the first equation and notice that on the left, $\frac{dR}{dt} = 2 e^{2t}$. Then look at the right side of the first equation: 
$$4.5R - 2.5W = 4.5e^{2t} - 2.5e^{2t} = 2e^{2t}$$
This equals $\frac{dR}{dt}$, so this pair of functions satisfies the first equation. In the second equation, $\frac{dW}{dt} = 2e^{2t}$ on the left; and on the right, 
$$7.5R - 5.5W = 7.5e^{2t} - 5.5e^{2t} = 2e^{2t}$$
Therefore the pair satisfies the second equation too. 

There is one obvious solution to this system: The pair of constant functions $R(t) = 0, W(t) = 0$. You can check that the derivative of each of these is zero, and when plugged into the right hand sides of each DE in the system the result is also zero. Solutions to a system of DE's where both functions are constant, are called **equilibrium solutions**. In this case, the equilibrium solution is not very interesting from a real-world perspective (it's an ecosystem with no rabbits and no wolves) but very often, equilibrium solutions are of great interest. 

### How do systems of DE's connect back to linear algebra? 

Going back to the earlier system: 

$$\begin{align}
\frac{dR}{dt} &= 4.5R - 2.5W \\
\frac{dW}{dt} &= 7.5R  - 5.5W
\end{align}$$

This looks a lot like a system of linear equations from earlier in our studies. And in fact we can use vectors and matrices to work with systems of DEs to great effect. 

Take the pair of functions $R(t), W(t)$ and package them into a vector $\mathbf{x}$: 

$$\mathbf{x}(t) = \begin{bmatrix} R(t) \\ W(t) \\ \end{bmatrix}$$

Then this vector $\mathbf{x}$ is a *vector function*, like we learned about in Calculus 3. You can think of it as a [parametric curve](https://math.libretexts.org/Bookshelves/Calculus/Map%3A_Calculus__Early_Transcendentals_(Stewart)/10%3A_Parametric_Equations_And_Polar_Coordinates/10.01%3A_Curves_Defined_by_Parametric_Equations). And in Calculus, we learned that we can take the derivative of such a function just by differentiating each of its component functions: 

$$\mathbf{x}'(t) = \begin{bmatrix} R'(t) \\ W'(t) \\ \end{bmatrix}$$

([Click here for a review.](https://www.khanacademy.org/math/multivariable-calculus/multivariable-derivatives/differentiating-vector-valued-functions/a/derivatives-of-vector-valued-functions)) So we can rewrite the system of DEs in **matrix form**: 

$$\begin{bmatrix} R'(t) \\ W'(t) \\ \end{bmatrix} = \begin{bmatrix} 4.5 & -2.5 \\ 7.5 & -5.5 \end{bmatrix} \begin{bmatrix} R(t) \\ W(t) \\ \end{bmatrix}$$

If you're having trouble seeing this, start at the end above, and multiply the matrix by the vector -- you'll get the right-hand sides of the original system. 

And if we define $\mathbf{x}(t) = \begin{bmatrix} R(t) \\ W(t) \\ \end{bmatrix}$, we can rewrite this even further as: 

$$\mathbf{x}' = \begin{bmatrix} 4.5 & -2.5 \\ 7.5 & -5.5 \end{bmatrix} \mathbf{x}$$

In other words, this *system* of DEs can be written as a *single* DE... of a vector function. 

**Important:** Note that this way of writing a system of DE's only works if the coefficients on the right-hand sides are **constant** (numbers). Later on, we'll examine systems of DE's where the coefficients are non-constant. 

### Another example

Consider the system of DE's

$$\begin{align}
\frac{dx}{dt} &= -2x + 3y \\ 
\frac{dy}{dt} &= 5y
\end{align}$$

**A non-equilibrium solution:** Let $x(t) = 3e^{5t}$ and $y(t) = 7e^{5t}$. To check that this is a solution, start by checking the first equation: 

$$\begin{align}
\frac{dx}{dt} &= 15e^{5t} \\ 
-2x + 3y &= -6e^{5t} + 21 e^{5t} = 15e^{5t}
\end{align}$$

So the first DE is satisfied. *For practice, check that the second DE is satisfied.*

**Putting this into matrix form:** Let $\mathbf{x} = [3e^{5t}, 7e^{5t}]^T$. Then the system becomes
$$\mathbf{x}' = \begin{bmatrix} -2 & 3 \\ 0 & 5 \end{bmatrix} \mathbf{x}$$

### What do solutions to systems of DE's look like? 

Solutions to systems of differential equations are pairs of functions $x(t), y(t)$, that share a common input variable $t$ which we usually think of as "time". You can plot these functions as parametric curves. For example, here's a plot of the parametric curve given by the functions $x(t)= t^2, y(t) = t^3-3t+3$ done using SymPy's parametric plot function. (The arguments at the end put labels on the axes, and are not required.) 

```python
t = var("t")
plot_parametric(t**2, t**3 - 3*t + 3, (t, -2,2), xlabel = 'x', ylabel = 'y')
```

Here, we've plotted the curve traced out in the $xy$ plane by $x$ and $y$ as time varies. Alternatively, we could have plotted both functions together, with time on the horizontal axis.

```python
plot(t**2, t**3 - 3*t + 3, (t,-2,2), xlabel = 't (Time)', ylabel='x and y')
```

Going back to the predator-prey system 

$$\begin{align}
\frac{dR}{dt} &= 4.5R - 2.5W \\
\frac{dW}{dt} &= 7.5R  - 5.5W
\end{align}$$

We found a non-equilibrium solution $R(t) = e^{2t}, W(t) = e^{2t}$. Let's see what this solution looks like in the $RW$-plane: 

```python
plot_parametric(exp(2*t), exp(2*t), (t,0,3), 
                xlabel = "R (rabbits)", 
                ylabel = "W (wolves)")
```

It's a straight line! That shouldn't be surprising, since if we dig some more into our Calculus toolbox, eliminating the parameter in the parametric equations would just give us $R = W$, which is the line through the origin with slope 1. 

A solution to a system of differential equations that traces out a straight line in the plane like this, is called a **straight line solution** (SLS) to the system. 

**Important fact:** When we finally learn how to *find* solutions to systems of DEs, we'll start by finding all the straight-line solutions. 

### Another example

Let's go back to 

$$\begin{align}
\frac{dx}{dt} &= -2x + 3y \\ 
\frac{dy}{dt} &= 5y
\end{align}$$

We found one solution to be $x(t) = 3e^{5t}$ and $y(t) = 7e^{5t}$. Is this a straight-line solution? 

```python
plot_parametric(3*exp(5*t), 7*exp(5*t), (t, 0, 1))
```

It looks like it. And again, you can eliminate the parameter: Notice $e^{5t} = x/3$, so $y = 7 \cdot x/3$ or $y = \frac{7}{3}x$ --- which is a staight line. 

Here's a *non*-straight line solution for this system: 

$$x(t) = 100e^{-2t} + 3e^{5t}, y(t) = 7e^{5t}$$

*For practice: Check that this pair of functions solves the system.* If you plot them in the $xy$-plane, you get: 

```python
plot_parametric(100*exp(-2*t) + 3*exp(5*t), 7*exp(5*t), (t, 0, 1))
```

Which definitely does *not* plot a straight line. 

### Second-order differential equations

Systems of DEs give us our primary tool for dealing with *second*-order differential equations. Here is an example: $y'' + y' + 6y = 0$. This is a *linear, homogeneous second order DE*. Equations of this type that are *first*-order have a straightforward solution technique; but what about second order? 

The trick is that we can turn this single, second-order equation into a *system of two first-order equations*, like so: Just let $u = y'$. Therefore $u' = y''$. And since $y'' + y' + 6y = 0$, it means $y'' = -y' - 6y$ which is the same as $u' = -u - 6y$. So we have the system

$$\begin{align}
\frac{du}{dt} &= -u-6y \\
\frac{dy}{dt} &= u
\end{align}$$

In matrix form, if $\mathbf{x} = [u(t), y(t)]^T$, then this system looks like
$$\mathbf{x}' = \begin{bmatrix} -1 & -6  \\ 1 & 0 \end{bmatrix} \mathbf{x}$$

Later, we will learn how to solve this system. (This one turns out to be rather... complex.)
