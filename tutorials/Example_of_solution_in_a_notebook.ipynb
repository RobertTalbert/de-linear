## Solution

We are given the rate at which Sonya can paint, and the problem is asking for the total amount that Sonya can paint. We're specifically interested in how much square footage Sonya can paint in three hours. To get from the rate at which she can paint to the total amount she can paint in three hours, we need to find: 

$$\int_0^3 v(t) \, dt$$

which in this case is

$$\int_0^3 (150-4t)  \, dt$$

The work in SymPy is below: 

```python
t = symbols('t')

integrate(150-4*t, (t, 0, 3))
```

So Sonya can paint 432 square feet in 3 hours. Now we just need to find out how much square footage is in each office. 

The first office is $12 \times 12$ with an 8-foot ceiling. So each wall is 12 feet wide by 8 feet high, which means it covers 96 square feet. And there are four of these walls, so there's a total of $96 \times 4 = 384$ square feet to paint. This is less than 432, so the answer to the first question is **YES**, Sonya can paint this room in 3 hours. 

The second office on the other hand is $14 \times 14$, so each wall has $14 \times 8 = 112$ square feet. There are four of these walls, giving a total of $112 \times 4 = 448$ square feet. This is greater than 432, so the second question is answered **NO**, Sonya cannot paint this entire room in three hours.
